```python
stock_no = ["0005", "0168", "0270", "0323", "0388", "0467", "0659", "0700", "0883", "0939", "1043", "1088", "1171", "1610", "2360", "2388"]

#===================================================================================stock no
news = {}
news_title = {}
news_content = {}
count = 0 

#===================================================================================variables
news['stock_info'] = []

#===================================================================================list
for stock in stock_no:
    URL = "https://hk.finance.yahoo.com/quote/"+ stock + ".HK?ltr=1"
    driver = webdriver.Chrome()
    driver.get(URL)
    Y = 2000
    for i in range(5):
        scrolling = f"window.scrollTo(0, {Y})"
        driver.execute_script(scrolling)
        Y += 2000
        sleep(1.5)
    subhtml = driver.page_source
    soup = bs(subhtml, "html.parser")
    driver.quit()
    
    if stock not in news:                                         
        news['stock_info'].append({"stock_no": stock})
    news['stock_info'][count]['stock_news'] = {}
    news['stock_info'][count]['stock_news']['stock_title'] = []
    news['stock_info'][count]['stock_news']['stock_contentnews'] = []
    
    b_list = soup.find_all("h3")
    for index, i in enumerate(b_list[5:]):
        title_result = i.text
        if "【" not in title_result:  # remove part of advertisements
            news['stock_info'][count]['stock_news']['stock_title'].append(f'{index} {title_result}')
    
    time_ = soup.find_all('div', class_="Cf")
    for index, i in enumerate(time_[3:]):
        content_result = i.text
        if "【" not in content_result:
            news['stock_info'][count]['stock_news']['stock_contentnews'].append(f'{index} {content_result}')
    sleep(2.0)
    count += 1
            
```

# Google search

```python
query = "Nokia 近年 做咩"

query = query.replace(' ', '+')
URL = f"https://google.com/search?q={query}"
google_link = []
google_title = []
driver = webdriver.Chrome()
driver.get(URL)
for i in range(5):
    try:
        python_button = driver.find_elements_by_xpath("//a[@id='pnnext']")[0]
        python_button.click()
        sleep(1.5)
        html = driver.page_source
        soup = bs(html, "html.parser")
        for g in soup.find_all('div', class_='g'):
            anchors = g.find_all('a')
            if anchors:
                link = anchors[0]['href']
                google_link.append(link)
                title = g.find('h3').text
                google_title.append(title)
    except:
        driver.quit()
        pass
driver.quit()
```
